# 真正的卷积

之前已经完成了使用模板匹配做好了单个模板的卷积，现在我们扩展成多个模板的情况；做到这里就有点像ViT做的事情了，我们大概要做的事情如下：

- 将输入的图片分割成多个patch
- 将每个patch进行layer norm得到卷积核权重
- 将卷积核与原图像进行卷积

> 这里假定输入的图片是480x480的，patch size是32x32，那么我们就有15x15=225个patch

## 获取图片

```python
from PIL import Image
```

```python
sails = Image.open("images/sails.png")
sails
```

```python
import numpy as np
sails_np = np.array(sails)
sails_np.shape
```

```python
# 切割图片
sails_np = sails_np[:480, :480, :]
sails_np.shape
```

```python
## 保存图片
sails = Image.fromarray(sails_np)
sails.save("images/sails_480.png")
```

```python
# 切割图片
sails_np = sails_np[:, :512, :]
sails_np.shape
```

```python
## 保存图片
sails = Image.fromarray(sails_np)
sails.save("images/sails_512.png")
```
