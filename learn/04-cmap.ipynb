```python
# 
single_channel_image = np.random.rand(192, 192)  # 

# viridis
cmap = plt.get_cmap('viridis')
rgb_image = cmap(single_channel_image)

#  (3, 192, 192)
rgb_image = rgb_image.transpose(2, 0, 1)

# 
plt.subplot(1, 2, 1)
plt.imshow(single_channel_image, cmap='gray')
plt.title('Single Channel Image')

# 
plt.subplot(1, 2, 2)
plt.imshow(rgb_image.transpose(1, 2, 0))  # imshow
plt.title('RGB Image')

# 
plt.show()
```
