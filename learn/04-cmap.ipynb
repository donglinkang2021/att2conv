```python
# 创建单通道图像数据（示例）
single_channel_image = np.random.rand(192, 192)  # 替换为您的实际图像数据

# 使用默认的颜色映射（viridis）将单通道图像转为三通道
cmap = plt.get_cmap('viridis')
rgb_image = cmap(single_channel_image)

# 调整形状以符合 (3, 192, 192)
rgb_image = rgb_image.transpose(2, 0, 1)

# 显示原始单通道图像
plt.subplot(1, 2, 1)
plt.imshow(single_channel_image, cmap='gray')
plt.title('Single Channel Image')

# 显示转换后的三通道图像
plt.subplot(1, 2, 2)
plt.imshow(rgb_image.transpose(1, 2, 0))  # 调整形状以适应imshow函数
plt.title('RGB Image')

# 显示图形
plt.show()
```
