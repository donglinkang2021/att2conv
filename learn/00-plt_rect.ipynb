```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

#  img  [x, y, width, height]
img = np.random.rand(100, 100, 3)  # 
anchor_box = [20, 30, 40, 50]  # 

# 
plt.imshow(img)

# Rectangle
rect = patches.Rectangle((anchor_box[0], anchor_box[1]), anchor_box[2], anchor_box[3], linewidth=2, edgecolor='r', facecolor='none')

# Rectangle
plt.gca().add_patch(rect)

# 
plt.show()
```
