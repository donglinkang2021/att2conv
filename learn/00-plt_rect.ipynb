```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# 假设你有一张图片 img 和一个表示锚框位置的矩形 [x, y, width, height]
img = np.random.rand(100, 100, 3)  # 生成一个随机的图片作为例子
anchor_box = [20, 30, 40, 50]  # 示例锚框位置

# 显示图片
plt.imshow(img)

# 创建一个Rectangle对象表示锚框
rect = patches.Rectangle((anchor_box[0], anchor_box[1]), anchor_box[2], anchor_box[3], linewidth=2, edgecolor='r', facecolor='none')

# 将Rectangle对象添加到图中
plt.gca().add_patch(rect)

# 显示图像
plt.show()
```
